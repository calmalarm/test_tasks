# Обобщённый AUC

https://contest.yandex.ru/contest/39184/problems

https://contest.yandex.ru/contest/39184/run-report/69728925/

```python
%%time
import numpy as np

file = open("input.txt")
n = int(file.readline())

a = np.empty((n,2))
t = np.empty(n)
y = np.empty(n)
for i in range(n):
    a[i] = file.readline().split()
    t[i] = float(a[i][0])
    y[i] = float(a[i][1])
file.close()

del file
del a

s = t.argsort()
t = t[s]
y = y[s]

classes = set(t)
N = len(classes)
class_change_ind = [0]
for i in range(n-1):
     if (t[i] < t[i+1]):
            class_change_ind.append(i+1)
class_change_ind.append(n)

a = []
l = np.empty(N, dtype=int)
for i in range(N):
    a.append(np.copy(y[class_change_ind[i]:class_change_ind[i+1]]))
    a[i].sort()
    l[i] = int(class_change_ind[i+1] - class_change_ind[i])

divider = 0.0
dividend = 0.0

for x in range(N-1):
    for w in range(x+1, N):
        divider += l[x] * l[w]
        for i in reversed(range(l[x])):
            for j in (range(l[w])):
                if (a[x][i] > a[w][j]):
                    dividend += 1.0
                else:
                    if(a[x][i] == a[w][j]):
                        dividend += 0.5
                    else:
                        break

res = 1 - dividend / divider
print(f"{res:2.6f}")
```
